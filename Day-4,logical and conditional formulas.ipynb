Logical and Conditional Formulas in Excel  <br>
    <h3 style="text-align: center; color: white; font-size: 15px; font-family: 'Arial', sans-serif;"></h3>
</div>

Logical formulas in Excel use functions to perform logical tests and return TRUE or FALSE. They're essential for decision-making, filtering data, and conditional calculations. Here are the **most commonly used logical formulas** in Excel:

---

###  1. `IF` Function

Performs a logical test and returns one value for TRUE and another for FALSE.

```excel
=IF(A1>10, "Yes", "No")
```

> Returns "Yes" if A1 is greater than 10, otherwise "No".

---

###  2. `AND` Function

Returns TRUE if **all conditions** are TRUE.

```excel
=AND(A1>5, B1<10)
```

> TRUE if A1 is greater than 5 **and** B1 is less than 10.

---

###  3. `OR` Function

Returns TRUE if **any condition** is TRUE.

```excel
=OR(A1>5, B1<10)
```

> TRUE if either A1 > 5 **or** B1 < 10 (or both).

---

###  4. `NOT` Function

Reverses the logical value: TRUE becomes FALSE, and vice versa.

```excel
=NOT(A1>10)
```

> TRUE if A1 is **not greater** than 10.

---

###  5. Nested `IF`

Handles multiple conditions.

```excel
=IF(A1>90, "A", IF(A1>80, "B", "C"))
```

> Grades based on score ranges.

---

###  6. `IF` with `AND` / `OR`

```excel
=IF(AND(A1>10, B1<5), "OK", "Not OK")
```

> Combines conditions in one formula.

---

### 7. `IFS` (Excel 2016+)

Simpler alternative to nested `IF`s.

```excel
=IFS(A1>90, "A", A1>80, "B", A1>70, "C")
```

---

###  8. `SWITCH` (Excel 2016+)

Compares one expression against multiple values.

```excel
=SWITCH(A1, "Red", 1, "Green", 2, "Blue", 3)
```

> Returns number based on color name.

---

###  9. `XOR` (Exclusive OR)

Returns TRUE if **only one** condition is TRUE.

```excel
=XOR(A1>10, B1>10)
```

> TRUE if **either** A1 or B1 is > 10, but not both.

---

###  Conditional Formulas 

**Conditional formulas** allow you to perform actions **based on certain conditions** — very useful in grading, pricing, status checking, etc.

---

###  1. Basic `IF` Condition

```excel
=IF(A1>100, "High", "Low")
```

* If A1 is greater than 100, result is **"High"**; otherwise, **"Low"**.

---

###  2. `IF` with Text

```excel
=IF(A1="Pass", "Congratulations", "Try Again")
```

* Checks if A1 contains "Pass".

---

###  3. `IF` with Multiple Conditions – `AND`

```excel
=IF(AND(A1>=50, B1="Pass"), "Qualified", "Not Qualified")
```

* Only returns "Qualified" if both conditions are true.

---

###  4. `IF` with `OR`

```excel
=IF(OR(A1="Absent", A1="NA"), "No Result", "Check Score")
```

* Handles multiple possible text values.

---

###  5. Nested `IF` – For Grades

```excel
=IF(A1>=90, "A", IF(A1>=80, "B", IF(A1>=70, "C", "Fail")))
```

* Assigns grade based on score.

---

###  6. Conditional Math

```excel
=IF(A1>100, A1*0.1, 0)
```

* If A1 > 100, gives 10% of A1; else returns 0.

---

###  7. Use with `ISBLANK` – Check for Empty Cells

```excel
=IF(ISBLANK(A1), "No Input", A1)
```

* Tells if a cell is empty.

---

###  8. `IFERROR` – Handle Errors

```excel
=IFERROR(A1/B1, "Error")
```

* Prevents errors like division by zero.

---

###  9. Conditional Formatting (Not a formula, but used conditionally)

To highlight cells:

* Go to **Home > Conditional Formatting > New Rule**
* Use a formula like:

```excel
=A1>100
```

To change color of cells > 100.
